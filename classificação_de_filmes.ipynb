## **Importando bibliotecas**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
plt.style.use('ggplot')
```

## **Visualizando o Dataset**

Primeiro iremos carregar só a base de dados dos filmes que contém gênero, id e título.

```python
filmes = pd.read_csv('/content/movies.dat', delimiter = '::')
filmes
```

Como essa base não possuí nome das colunas iremos definir abaixo.

```python
filmes.columns = ['ID', 'Título', 'Gênero']
filmes
```

Agora iremos carregar a base de dados de avaliações.

```python
avaliações = pd.read_csv('/content/ratings.dat', delimiter = '::')
avaliações
```

Esta base também não veio com o nome das colunas, iremos definir abaixo.

```python
avaliações.columns = ['Usuário', 'ID', 'Notas', 'Timestamp']
avaliações
```

Agora vamos juntar essas duas bases de dados. Ambas tem o ID do filme como uma coluna em comum, então podemos utiliza-la para mesclar os dados.

```python
nova = pd.merge(filmes, avaliações, on = ['ID', 'ID'])
nova
```

## **Analise**

```python
nova.Notas.value_counts()
```

```python
av = nova['Notas'].value_counts()
numeros = av.index
quantidade = av.values
fig = px.pie(nova, values=quantidade, names=numeros)
fig.show()
```

De acordo com o gráfico a maioria dos filmes são classificados positivamente com nota 8.

Agora daremos uma olhada nos filmes que obtiveram a nota máxima(10).

```python
nova2 = nova.query('Notas==10')
print(nova2['Título'].value_counts().head(10))
```

Podemos observar que o Coringa tem sido o queridinho do público. São 1479 notas 10.
